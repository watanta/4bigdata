# 機械学習用特徴量

## 線形補間したデータで作る

```python
width = 2
lag = 5
target = 10015

read_kilo = list(range(target-width,target+width+1))
data = pd.DataFrame()
for kilo in read_kilo:
    data[str(kilo)] = pd.read_csv("../data/kilo"+str(kilo)+"_interpolate.csv")["高低左"]
data.index = pd.to_datetime(pd.read_csv("../data/kilo"+str(kilo)+"_interpolate.csv")["date"])
data
```

```python
feature = pd.DataFrame()
for l in list(range(1,lag+1)):
    for kilo in read_kilo:
        feature.loc[:,str(kilo) + "lag"+str(l)] = data[str(kilo)].shift(-(lag-l))
feature.loc[:,"y"] = data[str(target)].shift(-lag)
feature.index = feature.index + offsets.Day(lag)
feature
```

```python
feature.to_csv("../data/"+str(target)+"_w"+str(width)+"_l"+str(lag)+"int.csv")
```
