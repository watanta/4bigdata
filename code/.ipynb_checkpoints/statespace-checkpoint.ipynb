```python
import pystan
from pystan import StanModel

import numpy as np
import numpy.random as rd
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import animation as ani
import matplotlib.cm as cm
import seaborn as sns
sns.set(style="whitegrid", palette="muted", color_codes=True)

from sklearn import cross_validation
from sklearn.metrics import mean_squared_error

import pickle
```

```python
kilo10000 = pd.read_csv("../data/kilo10000_interpolate.csv")
kilo10001 = pd.read_csv("../data/kilo10001_interpolate.csv")
kilo10002 = pd.read_csv("../data/kilo10002_interpolate.csv")
kilo10003 = pd.read_csv("../data/kilo10003_interpolate.csv")
kilo10004 = pd.read_csv("../data/kilo10004_interpolate.csv")
data = pd.concat((kilo10000["高低左"],kilo10001["高低左"],kilo10002["高低左"],kilo10003["高低左"],kilo10004["高低左"]), axis=1)
data.columns = ["10000","10001","10002","10003","10004"]
data.index = pd.to_datetime(kilo10000["date"])
data.head()
```

```python
# 訓練データ, テストデータに分割する
#https://note.nkmk.me/python-pandas-time-series-datetimeindex/
train = data["2017/4/1":"2017/12/31"]
test = data["2018/1/1":]
print ([d.shape for d in [train, test]])
```

# ローカルレベルモデル

```python
model = """
    data{
        int n; #サンプルサイズ
        vector[n] y;　#観測値
    }
    parameters{
        real muZero;
        vector[n] mu;
        vector[n] y_est;
        real<lower=0> sigmaW;
        real<lower=0> sigmaV;
    }
    model{
    #状態方程式
        mu[1] ~ normal(muZero, sqrt(sigmaW));
        for(i in 2:n){
            mu[i] ~ normal(mu[i-1], sqrt(sigmaW));
        }
    #観測方程式
    for(i in 1:n){
        y[i] ~ normal(mu[i], sqrt(sigmaV));
    }
    #yのサンプルを得たい
    for(i in 1:n){
        y_est[i] ~ normal(mu[i], sqrt(sigmaV));
    }
    }
"""
```

```python
n = train["10000"].shape[0]
y = train["10000"].values
```

```python
itr = 2000
warmup = 500
data = {"n":n, "y": y}
chains = 4

fit = pystan.stan(model_code=model, data=data, iter=itr, chains=chains, warmup=warmup)
fit
```

```python
model_name = "local_level"
with open(model_name + '.pkl', 'wb') as f:
    pickle.dump(fit, f)
```

```python
fit = pickle.load(open(model_name + '.pkl', 'rb'))
```

```python
la = fit.extract()
```

```python
##収束確認
param_name = "lp__" #確認するパラメータ
d = la[param_name]
sns.tsplot(d,   alpha=0.8, lw=.5)
```

```python
pred = la['mu'].mean(axis=0)
mu_lower, mu_upper = np.percentile(la['mu'], q=[2.5, 97.5], axis=0)
```

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(111)

ax1.plot(train["10000"].values)
ax1.plot(pred)
plt.fill_between(list(range(1,train["10000"].shape[0]+1)), mu_lower, mu_upper, alpha=0.3)
fig.suptitle("estimation_fo_mu kilo10000")
```

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(111)

ax1.plot(train["10000"].values)
ax1.plot(y_est)
plt.fill_between(list(range(1,train["10000"].shape[0]+1)), y_est_lower, y_est_upper, alpha=0.3)
fig.suptitle("estimation_of_y kilo10000")
```

```python
#予測値を得るためにはstanモデル自体にパラメータとして書かなければいけない
model = """
    data{
        int n; #トレインサンプルサイズ
        vector[n] y;　#観測値
        int n_pred; #テストサンプルサイズ
    }
    parameters{
        real muZero;
        vector[n] mu;
        real<lower=0> sigmaW;
        real<lower=0> sigmaV;
        vector[n_pred] y_pred;
        vector[n_pred] mu_pred;
        vector[n] y_est;
    }
    model{
    #状態方程式
        mu[1] ~ normal(muZero, sqrt(sigmaW));
        for(i in 2:n){
            mu[i] ~ normal(mu[i-1], sqrt(sigmaW));
        }
        mu_pred[1] ~ normal(mu[n], sqrt(sigmaW));
        for(i in 2:n_pred){
            mu_pred[i] ~ normal(mu_pred[i-1], sqrt(sigmaW));
        }
    #観測方程式
    for(i in 1:n){
        y[i] ~ normal(mu[i], sqrt(sigmaV));
    }
    for(i in 1:n_pred){
        y_pred[i] ~ normal(mu_pred[i], sqrt(sigmaV));
    }
    #yのサンプルを得たい
    for(i in 1:n){
        y_est[i] ~ normal(mu[i], sqrt(sigmaV));
    }

    }
"""
```

```python
n = train["10000"].shape[0]
y = train["10000"].values
n_pred = test["10000"].shape[0]
```

```python
itr = 3000
warmup = 500
data = {"n":n, "y": y, "n_pred":n_pred}
thin = 2
chains = 4

fit = pystan.stan(model_code=model, data=data, iter=itr, chains=chains, warmup=warmup, thin = thin)
fit
```

```python
model_name = "local_level_pred"
with open(model_name + '.pkl', 'wb') as f:
    pickle.dump(fit, f)
```

```python
fit = pickle.load(open(model_name + '.pkl', 'rb'))
```

```python
la = fit.extract()
```

# ローカルトレンドモデル

```python
model = """
    data{
        int n; #サンプルサイズ
        vector[n] y;　#観測値
    }
    parameters{
        real muZero; #レベル初期値
        real trendZero; #トレンド初期値
        vector[n] mu;
        vector[n] trend;
        real<lower=0> sigmaW_mu; # レベルの分散
        real<lower=0> sigmaW_tr; # トレンドの分散
        real<lower=0> sigmaV; #観測雑音
        vector[n] y_est; #yの推定値
    }
    model{
    #状態方程式
        mu[1] ~ normal(muZero, sqrt(sigmaW_mu));
        trend[1] ~ normal(trendZero, sqrt(sigmaW_tr));
        for(i in 2:n){
            mu[i] ~ normal(mu[i-1] + trend[i-1], sqrt(sigmaW_mu));
            trend[i] ~ normal(trend[i-1], sqrt(sigmaW_tr));
        }
    #観測方程式
    for(i in 1:n){
        y[i] ~ normal(mu[i], sqrt(sigmaV));
    }
    #yのサンプルを得たい
    for(i in 1:n){
        y_est[i] ~ normal(mu[i], sqrt(sigmaV));
    }
    }
    
"""
```

```python
n = train["10000"].shape[0]
y = train["10000"].values
```

```python
itr = 5000
warmup = 500
thin = 4
data = {"n":n, "y": y}
chains = 4

fit = pystan.stan(model_code=model, data=data, iter=itr, chains=chains, warmup=warmup, thin=thin)
fit
```

```python
model_name = "local_trend"
with open(model_name + '.pkl', 'wb') as f:
    pickle.dump(fit, f)
```

```python
fit = pickle.load(open(model_name + '.pkl', 'rb'))
```

# ARMA(2,1)

```python
model = """
    data{
        int n; #トレインサンプルサイズ
        vector[n] y;　#観測値
    }
    parameters{
        real xZero1; #状態初期値
        real xZero2; #状態初期値
        vector[2] AR; #AR成分
        real MA; #MA成分
        matrix[n,2] x ;#状態  
        real<lower=0> sigma_W; #状態誤差
        
    }
    model{
    #状態方程式
     //   x[1,1] ~ normal(AR[1]*xZero1 + xZero2, sqrt(sigma_W));
     //   x[1,2] ~ normal(AR[2]*xZero1, MA*sqrt(sigma_W));
     
        x[1,1] ~ normal(AR[1], sqrt(sigma_W));
        x[1,2] ~ normal(AR[2], MA*sqrt(sigma_W));
     
     
        for(i in 2:n){
        x[i,1] ~ normal(AR[1]*x[i-1,1] + x[i-1,2], sqrt(sigma_W));
        x[i,2] ~ normal(AR[2]*x[i-1,1], MA*sqrt(sigma_W));
     
     
     
       }
        
    #観測方程式
    for(i in 1:n){
        y[i] ~ normal(x[i,1],1);
    }
    }
    
generated quantities {
    #yのサンプルを得たい
    vector[n] y_est; #yの推定値
    for(i in 1:n){
        y_est[i] = normal_rng(x[i,1],0);
    }
    }
    
"""
```

```python
n = train["10000"].shape[0]
y = train["10000"].values
```

```python
itr = 5000
warmup = 500
thin = 4
data = {"n":n, "y": y}
chains = 4
init_r = 1

fit = pystan.stan(model_code=model, data=data, iter=itr, chains=chains, warmup=warmup, thin=thin, init_r=init_r )
fit
```

# AR(2)

```python
model = """
    data{
        int n; #トレインサンプルサイズ
        vector[n] x1;　#観測値で状態
    }
    parameters{
        real xZero1; #状態初期値
        real xZero2; #状態初期値
        vector[2] AR; #AR成分
        //vector[n] x2; #状態
        
        real<lower=0> sigma_W; #状態誤差
        
    }
    
    transformed parameters{
        vector[n] x2; #状態
        x2[1] = xZero1;
        for(i in 2:n){
        x2[i] = x1[i-1];
        }
    }
    model{
    
    #状態方程式
        x1[1] ~ normal(AR[1]*xZero1 + AR[2]*xZero2, sqrt(sigma_W));
        

        for(i in 2:n){
        x1[i] ~ normal(AR[1]*x1[i-1] + AR[2]*x2[i-1], sqrt(sigma_W));
        
        //x1[1] ~ normal(AR[1]*x1[i-1] + AR[2]*x1[i-2], sqrt(sigma_W));
        }
       }
       

       
    generated quantities {
        vector[n] y;
        for(i in 1:n){
        y[i] = normal_rng(AR[1]*x1[i] + AR[2]*x2[i], sqrt(sigma_W));
        }
    }
    
"""
```

```python
n = train["10000"].shape[0]
x1 = train["10000"].values
```

```python
itr = 5000
warmup = 500
thin = 4
data = {"n":n, "x1": x1}
chains = 4
init_r = 1

fit = pystan.stan(model_code=model, data=data, iter=itr, chains=chains, warmup=warmup, thin=thin, init_r=init_r )
fit
```

```python
model_name = "ar2"
with open(model_name + '.pkl', 'wb') as f:
    pickle.dump(fit, f)
```

# AR2+ローカルレベル

```python
model = """

    data{
        int n; #サンプルサイズ
        vector[n] y;　#観測値
    }
    
    parameters{
        #ローカルレベルモデルの状態
        real muZero;
        vector[n] mu;
        
        #AR2モデルの状態
        real x1Zero;
        real x2Zero;
        vector[n] x1;
        
        #ローカルレベルモデルの状態の分散
        real<lower=0> sigmaW_mu;
        
        #AR2モデルの状態の分散
        real<lower=0> sigmaW_ar;
        
        #観測誤差
        real<lower=0> sigmaV;
        
        #AR成分
        vector[2] AR;
    }
    
    transformed parameters{
        #AR2モデルの状態
        vector[n] x2; 
        x2[1] = x1Zero;
        for(i in 2:n){
        x2[i] = x1[i-1];
        }
    }
    
    model{
    #状態方程式
        mu[1] ~ normal(muZero, sqrt(sigmaW_mu));
        x1[1] ~ normal(AR[1]*x1Zero + AR[2]*x2Zero, sqrt(sigmaW_ar));
        for(i in 2:n){
            mu[i] ~ normal(mu[i-1], sqrt(sigmaW_mu));
            x1[i] ~ normal(AR[1]*x1[i-1] + AR[2]*x2[i-1], sqrt(sigmaW_ar));
        }
        
    #観測方程式
    for(i in 1:n){
        y[i] ~ normal(x1[i] + mu[i], sqrt(sigmaV));
        }
    }
    
    generated quantities{
    #yのサンプルを得たい
    vector[n] y_est;
    for(i in 1:n){
        y_est[i] = normal_rng(x1[i] + mu[i], sqrt(sigmaV));
    }
    }
    

"""
```

```python
n = train["10000"].shape[0]
y = train["10000"].values
```

```python
itr = 10000
warmup = 3000
thin = 4
data = {"n":n, "y": y}
chains = 4
init_r = 1

fit = pystan.stan(model_code=model, data=data, iter=itr, chains=chains, warmup=warmup, thin=thin, init_r=init_r )
fit
```

```python
model = """

    data{
        int n; #サンプルサイズ
        vector[n] y;　#観測値
        int n_pred;
    }
    
    parameters{
        #ローカルレベルモデルの状態
        real muZero;
        vector[n] mu;
        
        #AR2モデルの状態
        real x1Zero;
        real x2Zero;
        vector[n] x1;
        
        #ローカルレベルモデルの状態の分散
        real<lower=0> sigmaW_mu;
        
        #AR2モデルの状態の分散
        real<lower=0> sigmaW_ar;
        
        #観測誤差
        real<lower=0> sigmaV;
        
        #AR成分
        vector[2] AR;
        
        
    }
    
    transformed parameters{
        #AR2モデルの状態
        vector[n] x2; 
        x2[1] = x1Zero;
        for(i in 2:n){
        x2[i] = x1[i-1];
        }
    }
    
    model{
    #状態方程式
        mu[1] ~ normal(muZero, sqrt(sigmaW_mu));
        x1[1] ~ normal(AR[1]*x1Zero + AR[2]*x2Zero, sqrt(sigmaW_ar));
        for(i in 2:n){
            mu[i] ~ normal(mu[i-1], sqrt(sigmaW_mu));
            x1[i] ~ normal(AR[1]*x1[i-1] + AR[2]*x2[i-1], sqrt(sigmaW_ar));
        }
        
    
        
    #観測方程式
    for(i in 1:n){
        y[i] ~ normal(x1[i] + mu[i], sqrt(sigmaV));
        }
    }
    
    
    
    
    generated quantities{
    
    
    #yのサンプルを得たい
        vector[n] y_est;
    
        vector[n_pred] mu_pred;
        vector[n_pred] AR_pred;
        vector[n_pred] y_pred;
        vector[n_pred] x1_pred;
        vector[n_pred] x2_pred;
    
        for(i in 1:n){
            y_est[i] = normal_rng(x1[i] + mu[i], sqrt(sigmaV));
        }
    
    #予測
        
        mu_pred[1] = normal_rng(mu[n], sqrt(sigmaW_mu));
        x1_pred[1] = normal_rng(AR[1]*x1[n] + AR[2]*x2[n], sqrt(sigmaW_ar));
        for(i in 2:n_pred){
            mu_pred[i] = normal_rng(mu_pred[i-1], sqrt(sigmaW_mu));
            x1_pred[i] = normal_rng(AR[1]*x1_pred[i-1] + AR[2]*x2_pred[i-1], sqrt(sigmaW_ar));
        
        }
    
        for(i in 1:n_pred){
        y_pred[i] = normal_rng(x1_pred[i] + mu_pred[i], sqrt(sigmaV));
        }
    }
        
        
    
    

"""
```
